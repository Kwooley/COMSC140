```python
import matplotlib.pyplot as plt
x_coords = [0, 1, 2, 3, 4]
y_coords = [0, 3, 1, 5, 2]

plt.plot(x_coords, y_coords)
plt.show()
```

## Adding a Title, Axis Labels, and a Grid

```python

plt.title("Example 1 data")
plt.xlabel("X Axis")
plt.xlabel("Y Axis")
plt.grid(True)

plt.plot(x_coords, y_coords)
plt.show()
```

## Tick mark's label 

```python
import matplotlib.pyplot as plt
x_coords= [0,1,2,3,4]
y_coords=[1,2,3,4,5]
plt.xticks(x_coords, ['2022', '2021', '2020', '2019', '2018'])
plt.yticks(y_coords, ['$1', '$2', '$3', '$4', '$5'])
plt.plot(x_coords, y_coords, marker='D')
plt.show()
```

## Bar Chart

```python

x = [0, 10, 20, 30, 40]
y = [100, 200, 300, 200, 100]

bar_width = 5
plt.bar(x, y, bar_width)
plt.show
```

## Pie Chart

```python
import matplotlib.pyplot as plt

sales = [100, 400, 300, 600]
chart_label = ['2019', '2020', '2021', '2022']

plt.pie(sales, labels=chart_label, colors = ('w', 'b', 'g', 'c'))
plt.title('Pie Chart')
plt.show()
```

## Exercise 17: Stacked Bar Chart 2

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
students = ['John', 'Doe', 'Mary', 'Bill']
x = [1,2,3,4]
math = [80, 90, 50, 80]
english = [60, 80, 40, 50]

ax.bar(students, math)
ax.bar(students, english, bottom=math)
plt.show()
```

## Exercise 18: Grouped bar chart with a label

```python
import matplotlib.pyplot as plt
import numpy as np


label = ['Math', 'Physic', 'English', 'Algebra']
x = np.arange(len(label))
y1 = [100, 90, 100, 90]
y2 = [100, 100, 100, 90]
width = 0.35

fig, ax = plt.subplots()
rect1 = ax.bar(x-width/2, y1, width, label='John')
rect2 = ax.bar(x+width/2, y2, width, label='Doe')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.legend()
ax.set_xticks(x, label)
# plt.xticks(x, label)


ax.bar_label(rect1, padding=3)
ax.bar_label(rect2, padding=3)

plt.show()
```
