```python
from matplotlib import pyplot as plt

x = [1,2,3,4,5]
plt.plot(x)
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 20, 10] 

plt.bar(x, y, width=0.5)
plt.show()
```

```python
plt.xlabel(" X values")
plt.ylabel(" Scores")
plt.title(" Sample Bar Graph")
label = ['Math', 'English', 'Comp', 'Physics', 'Alg']
plt.bar(x, y, tick_label = label, color = ['green', 'blue'])
```
